## Libraries

```python
#Standard libraries
import pandas as pd
import numpy as np
```

## Loading in Data

```python
#Loading and checking both years
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz', 
                       low_memory = False)
df_2000.head()
```

```python
df_2000.shape
```

- 1222 Movies from 2000

```python
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz',
                       low_memory = False)
df_2001.head()
```

```python
df_2001.shape
```

- 1305 Movies from 2001

```python
#Concatenate to one dataframe
dfs = [df_2000, df_2001]
merged_df = pd.concat(dfs)
merged_df.head()
```

```python
merged_df.shape
```

- 2527 Movies total after merging 2000 and 2001
    - This is the number we expect so they merged properly

```python
#Saving the merged dataframe to a csv file
merged_df.to_csv('Data/tmdb_results_combined.csv.gz', compression='gzip', index=False)
```

## Movies with Valid Financial Information

```python
#Filling missing values with "0" so we can explore
merged_df['budget'] = merged_df['budget'].fillna(0)
merged_df['revenue'] = merged_df['revenue'].fillna(0)

merged_df.head()
```

```python
#Finding how many movies have some valid financial information
valid_fin_info = merged_df.loc[((merged_df['budget'] > 0) |
                                (merged_df['revenue'] > 0))]
valid_fin_info.shape
```

- There is 632 movies with at least some valid financial info for years 2000 and 2001

```python
#Excluding Movies with 0 for budget AND revenue for remaining analysis
filter_df = merged_df.loc[~((merged_df['budget'] == 0.0) & 
                            (merged_df['revenue'] == 0.0))]
```

## Movies per Cert Category

```python
filter_df['certification'].value_counts()
```

## Average Revenue per Cert Category

```python
filter_df.groupby('certification')['revenue'].mean()
```

## Average Budget per Cert Category

```python
filter_df.groupby('certification')['budget'].mean()
```
