# Loading in 2002-2019 Data

```python
os.listdir(FOLDER+'2002-2019')
```

```python
q = FOLDER+"/**/final_*.csv.gz"
print(q)
file_list = sorted(glob.glob(q, recursive=True))
file_list
```

```python
df = pd.concat([pd.read_csv(f, lineterminator='\n') for f in file_list])
df
```

```python
#Dropping first row with null index
df = df.loc[df['imdb_id']!='0']
df.head()
```

```python
#Reseting index
df = df.reset_index(drop=True)
df.head()
```

```python
df.to_csv(FOLDER+'complete_tmdb_data.csv.gz', compression='gzip', index=False)
```
